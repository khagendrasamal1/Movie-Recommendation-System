# Feature Selection

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genre', 'vote_average', 'release_date', 'overview']]
```

```python
movies
```

```python
movies['tags'] = movies['genre']+movies['overview']
```

```python
movies
```

```python
new_data = movies.drop(columns = ['overview'])
```

```python
new_data
```

# Data Transformation

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

# Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title'] == "The Godfather"].index[0]
```

```python
new_data.info()
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[1:5]:
    print(new_data.iloc[i[0]].title)
    print(new_data.iloc[i[0]].genre)
    print(new_data.iloc[i[0]].vote_average)
    print(new_data.iloc[i[0]].release_date)
```

```python
def recommend(movies):
    index = new_data[new_data['title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[1:5]:
        print(new_data.iloc[i[0]].title)
        print(new_data.iloc[i[0]].genre)
        print(new_data.iloc[i[0]].vote_average)
        print(new_data.iloc[i[0]].release_date)
```

```python
recommend("The Shawshank Redemption")
```

```python
recommend("Dune")
```
